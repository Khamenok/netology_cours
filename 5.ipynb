```python
import pandas as pd

ratings_path = r"C:\Users\mv.khamenok\Desktop\Документы\Netology\Py\ml-latest-small\ratings.csv"
ratings = pd.read_csv(ratings_path)

user_stat = ratings.groupby("userId").agg(
    ratings_cnt=("rating", "count"),
    min_ts=("timestamp", "min"),
    max_ts=("timestamp", "max")
).reset_index()

user_stat["lifetime_sec"] = user_stat["max_ts"] - user_stat["min_ts"]

active_users = user_stat[user_stat["ratings_cnt"] > 100]

mean_lifetime_sec = active_users["lifetime_sec"].mean()
mean_lifetime_days = mean_lifetime_sec / 86400

print("Пользователи, которые выставили более 100 оценок: ", len(active_users))
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
```

```python
revenue_table = (client_base[['client_id']]
                 .merge(rzd, on='client_id', how='left')
                 .merge(auto, on='client_id', how='left')
                 .merge(air, on='client_id', how='left')
                 .fillna(0))

revenue_table
```

```python
revenue_with_address = revenue_table.merge(client_base, on='client_id', how='left')
revenue_with_address = revenue_with_address[['client_id', 'address', 'rzd_revenue', 'auto_revenue', 'air_revenue']]
revenue_with_address
```
